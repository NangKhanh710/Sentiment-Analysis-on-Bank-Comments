# 3.Query ID of bank on GG map

```python
place_name = vietinbank + techcombank + agribank + acb + vpbank + mbbank + vib + hdbank + tpbank + bidv
#techcombank + mbbank + acb + hdbank + vib + bidv + vpbank + tpbank + agribank
place_ida = []
for i in place_name:
  place_result = gmaps.places(i)
  place_id_o = place_result['results'][0]['place_id']
  place_ida.append(place_id_o)

len(place_ida)
```

# 4. Export Data

```python
place_ids = place_ida
fields = ['name', 'rating', 'reviews']


df = pd.DataFrame(columns=['place_name', 'review_text', 'review_rating'])


for place_id in place_ids:

    place = gmaps.place(place_id=place_id, fields=fields, language='vi')


    place_name = place['result']['name']
    reviews = place['result']['reviews']


    for review in reviews:
        df = df.append({'place_name': place_name, 'review_text': review['text'], 'review_rating': review['rating']}, ignore_index=True)
```
